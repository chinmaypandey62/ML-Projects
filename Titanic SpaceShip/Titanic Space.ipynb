```python
cabin = train_data['Cabin'].str.split('/', expand = True)
```

```python
cabin_test = test_data['Cabin'].str.split('/', expand = True)
```

```python
cabin.rename(columns={0: 'Deck', 1: 'Num', 2: 'Side'}, inplace=True)
```

```python
cabin_test.rename(columns={0: 'Deck', 1: 'Num', 2: 'Side'}, inplace=True)
```

```python
# prompt: convert the categorical cols deck and side to numeric values

from sklearn.preprocessing import LabelEncoder

# Create a LabelEncoder object
label_encoder = LabelEncoder()

# Fit and transform the 'Deck' column
cabin['Deck'] = label_encoder.fit_transform(cabin['Deck'])

# Fit and transform the 'Side' column
cabin['Side'] = label_encoder.fit_transform(cabin['Side'])

cabin.head()
```

```python
# prompt: convert the categorical cols deck and side to numeric values

# Create a LabelEncoder object
label_encoder = LabelEncoder()

# Fit and transform the 'Deck' column
cabin_test['Deck'] = label_encoder.fit_transform(cabin_test['Deck'])

# Fit and transform the 'Side' column
cabin_test['Side'] = label_encoder.fit_transform(cabin_test['Side'])

cabin_test.head()
```

```python
# prompt: remove the cabin col from the train_data and append the cabin dataframe in the train_data

train_data = train_data.drop('Cabin', axis=1)
train_data = pd.concat([train_data, cabin], axis=1)
train_data.head()
```

```python
test_data = test_data.drop('Cabin', axis=1)
test_data = pd.concat([test_data, cabin_test], axis=1)
test_data.head()
```

```python
# prompt: convert the categorical cols homeplanet, destination, vip, transported to numeric values in train_data

# Create a LabelEncoder object
label_encoder = LabelEncoder()

# Iterate through the specified columns and convert them to numeric
for column in ['HomePlanet', 'Destination', 'VIP', 'Transported', 'CryoSleep']:
  if column in train_data.columns:
    # Handle missing values by filling with a new category (e.g., 'Unknown')
    train_data[column].fillna('Unknown', inplace=True)
    # Convert the column to string type before encoding to ensure uniformity
    train_data[column] = train_data[column].astype(str)
    train_data[column] = label_encoder.fit_transform(train_data[column])

train_data.head()
```

```python
# Iterate through the specified columns and convert them to numeric
for column in ['HomePlanet', 'Destination', 'VIP', 'Transported', 'CryoSleep']:
  if column in test_data.columns:
    # Handle missing values by filling with a new category (e.g., 'Unknown')
    test_data[column].fillna('Unknown', inplace=True)
    # Convert the column to string type before encoding to ensure uniformity
    test_data[column] = test_data[column].astype(str)
    test_data[column] = label_encoder.fit_transform(test_data[column])

test_data.head()
```

```python
train_data.fillna(train_data.mode().iloc[0], inplace=True)
```

```python
# The model to be tested for this problem will be KNN Classifier
```

```python
# We will first provide the weights to the features in the table.
# This will help KNN Classifier to decide how much priority should be given to the feature
```

```python
feature_importance = xgb_model.feature_importances_
```

```python
# Normalization: We divide the values with the largest value available in the feature
# Dividing like this shrinks the range between 0 to 1.
feature_importance /= feature_importance.max()
```

```python
X_train_weighted = X_train * feature_importance
X_test_weighted = X_test * feature_importance
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.pipeline import Pipeline
```

```python
standard_scaler = StandardScaler()
```

```python
# Train KNN on weighted features
knn = KNeighborsClassifier(n_neighbors=5, metric="euclidean")

# Create a pipeline
pipeline = Pipeline([("scaler", standard_scaler), ("knn", knn)])
pipeline.fit(X_train_weighted, y_train)
```

```python
accuracy = pipeline.score(X_test_weighted, y_test)
print(f"KNN Accuracy with Feature Weights: {accuracy:.4f}")
```
